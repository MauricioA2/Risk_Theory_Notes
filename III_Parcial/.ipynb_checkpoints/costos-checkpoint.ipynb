```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parámetros del modelo
costos_fijos = 100000  # Costos fijos en USD
costo_variable_unitario = 50  # Costo por unidad de producción en USD
```

```python
# Rango de producción (cantidad producida)
produccion = np.arange(0, 10001, 100)  # De 0 a 10,000 unidades, con incrementos de 100
```

```python
# Calcular costos totales y costos unitarios
costos_totales = costos_fijos + costo_variable_unitario * produccion
```

```python
# Evitar la división por cero y ajustar los costos unitarios
# Cuando la producción es cero, el costo unitario se considera muy alto, no infinito
costos_unitarios = np.divide(costos_totales, produccion, where=produccion != 0)  # Dividir solo cuando la producción no es cero
```

```python
# Para la producción cero, establecer un valor grande para el costo unitario (en lugar de infinito)
costos_unitarios[produccion == 0] = costos_fijos  # En este caso, costo unitario sería igual a los costos fijos
```

```python
# Mostrar algunos resultados
for i in range(0, len(produccion), 1000):
    print(f"Producción: {produccion[i]} unidades | Costo Total: ${costos_totales[i]:,.2f} | Costo Unitario: ${costos_unitarios[i]:,.2f}")
```

```python
# Graficar los resultados
plt.figure(figsize=(15, 5))

# Graficar Costo Total
plt.plot(produccion, costos_totales, label="Costo Total", color='blue', linewidth=2)

# Graficar Costo Unitario
plt.plot(produccion, costos_unitarios, label="Costo Unitario", color='red', linestyle='--', linewidth=2)

# Etiquetas y título
plt.title("Modelado de Costos en Función del Volumen de Producción")
plt.xlabel("Cantidad de Producción (Unidades)")
plt.ylabel("Costo (USD)")
plt.legend()

# Mostrar la gráfica
plt.grid(True)
plt.show()
```
