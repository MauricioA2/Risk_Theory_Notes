```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parámetros del modelo
beneficio_neto_anual = 1000000  # Beneficio neto anual proyectado (en USD)
payout_ratio = 0.44  # Porcentaje de beneficios pagados como dividendos
acciones_en_circulacion = 5000  # Número de acciones en circulación
```

```python
# Periodo de tiempo (5 años)
años = np.arange(1, 6)
```

```python
# Cálculo de dividendos anuales
dividendos_totales = beneficio_neto_anual * payout_ratio  # Dividendos totales anuales
dividendos_por_accion = dividendos_totales / acciones_en_circulacion  # Dividendos por acción
```

```python
# Proyección de dividendos a lo largo de los años
dividendos_totales_años = dividendos_totales * np.ones_like(años)
dividendos_por_accion_años = dividendos_por_accion * np.ones_like(años)
```

```python
# Mostrar los resultados
print(f"Beneficio neto anual: ${beneficio_neto_anual}")
print(f"Tasa de pago de dividendos (Payout Ratio): {payout_ratio * 100}%")
print(f"Dividendos totales anuales: ${dividendos_totales:.2f}")
print(f"Dividendos por acción: ${dividendos_por_accion:.2f}")
print("\nProyección de dividendos por año:")
for i in range(len(años)):
    print(f"Año {años[i]}: Dividendos totales = ${dividendos_totales_años[i]:,.2f}, Dividendos por acción = ${dividendos_por_accion_años[i]:,.2f}")
```

```python
# Graficar los resultados
plt.figure(figsize=(15, 5))
# Graficar Dividendos Totales
plt.plot(años, dividendos_totales_años, label="Dividendos Totales", color='blue', marker='o', linestyle='-', linewidth=3)
# Graficar Dividendos por Acción
plt.plot(años, dividendos_por_accion_años, label="Dividendos por Acción", color='green', marker='x', linestyle='--', linewidth=3)
# Etiquetas y título
plt.title("Proyección de Dividendos por Año")
plt.xlabel("Años")
plt.ylabel("Monto (USD)")
plt.legend()
# Mostrar la gráfica
plt.grid(True)
plt.show()
```
