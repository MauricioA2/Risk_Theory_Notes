```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Par치metros del problema
u_max = 1000  # Capital inicial m치ximo
c = 10       # Tasa de ingreso por primas
beta = 20    # Media de las reclamaciones (1/beta = par치metro de exponencial)
lambda_rate = 0.5  # Tasa de llegada de reclamaciones (Poisson)
delta_u = 1  # Incremento para la discretizaci칩n
```

```python
# Crear una lista para almacenar la probabilidad de ruina
psi = np.zeros(u_max + 1)  # Inicializamos en 0 para todos los valores de u
```

```python
# Calcular psi(u) iterativamente

for u in range(1, u_max + 1):
    integral_sum = 0
    for x in range(0, u):
        integral_sum += (1 / (c * beta)) * np.exp(-x / beta) * psi[u - x] * delta_u
    psi[u] = 1 - integral_sum
```

```python
# Gr치fica de la probabilidad de ruina
plt.figure(figsize=(15, 5))
plt.plot(range(u_max + 1), psi, label="Probabilidad de Ruina $\\psi(u)$", color='blue')
plt.title("Probabilidad de Ruina vs Capital Inicial")
plt.xlabel("Capital Inicial ($u$)")
plt.ylabel("Probabilidad de Ruina $\\psi(u)$")
plt.grid()
plt.legend()
plt.show()
```
