```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parámetros
n_periodos = 100  # 100 períodos (puede ser 100 años, 100 trimestres, etc.)
media_frecuencia = 5  # Promedio de siniestros por periodo
shape_severidad = 2  # Parámetro de forma de la distribución Gamma
scale_severidad = 1000  # Escala de la severidad
```

```python
# Simulando la frecuencia de siniestros (Poisson)
frecuencia = np.random.poisson(media_frecuencia, n_periodos)

# Simulando la severidad de los siniestros (Gamma)
severidad = np.random.gamma(shape_severidad, scale_severidad, n_periodos)

# Calculando el total de siniestros por período
total_siniestros = frecuencia * severidad
```

```python
# Graficar la evolución de los siniestros a lo largo de los períodos
plt.figure(figsize=(15,5))
plt.plot(np.arange(n_periodos), total_siniestros, label='Total Siniestros por Período')
plt.xlabel('Período')
plt.ylabel('Monto Total de Siniestros')
plt.title('Simulación de Siniestros con Horizonte Temporal Largo')
plt.legend()
plt.show()
```
