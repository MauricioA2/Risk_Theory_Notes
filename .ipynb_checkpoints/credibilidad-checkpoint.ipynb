# Credibilidad 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, poisson
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Parámetros de la distribución Gamma a priori
alpha_prior = 2  # Forma (shape)
beta_prior = 1   # Tasa (rate)
# Nuevos datos observados
x_observado = 3  # Número de siniestros observados
```

```python
# Actualizar la distribución a posteriori
alpha_posterior = alpha_prior + x_observado  # Forma de la distribución posterior
beta_posterior = beta_prior + 1  # Tasa de la distribución posterior
```

```python
# Media a priori y a posteriori
media_prior = alpha_prior / beta_prior
media_posterior = alpha_posterior / beta_posterior
```

```python
# Imprimir los resultados
print(f"Media a priori: {media_prior}")
print(f"Media a posteriori (actualizada): {media_posterior}")
```

```python
# Visualización de las distribuciones
lambdas = np.linspace(0, 10, 500)

# Distribución a priori
prior_pdf = gamma.pdf(lambdas, alpha_prior, scale=1/beta_prior)

# Distribución a posteriori
posterior_pdf = gamma.pdf(lambdas, alpha_posterior, scale=1/beta_posterior)

# Gráfica de las distribuciones
plt.figure(figsize=(15, 5))
plt.plot(lambdas, prior_pdf, label="Distribución a Priori", color="blue")
plt.plot(lambdas, posterior_pdf, label="Distribución a Posteriori", color="red", linestyle="--")
plt.title("Distribución a Priori y a Posteriori de $\lambda$")
plt.xlabel("Tasa de siniestros ($\lambda$)")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.grid(True)
plt.show()
```

# Credibilidad 2

```python
data = pd.read_excel("Ejemplo sieniestros.xlsx")
```

```python
data.head()
```

```python
# Calcular la media de los siniestros individuales y del mercado
media_individual = data['Datos Individuales'].mean()
media_mercado = data['Datos del Mercado'].mean()
```

```python
# Definir el número de datos individuales y un parámetro K (basado en la teoría de credibilidad de Bühlmann)
n = len(data['Datos Individuales'])  # Número de observaciones individuales
K = 10  # Parámetro K de la teoría de credibilidad, ajustable según el análisis actuarial
```

```python
# Calcular el factor de credibilidad Z
Z = n / (n + K)
```

```python
# Calcular la prima creíble
prima_creible = Z * media_individual + (1 - Z) * media_mercado
```

```python
# Imprimir los resultados
print(f"Media de siniestros individual: {media_individual:.2f}")
print(f"Media de siniestros del mercado: {media_mercado:.2f}")
print(f"Factor de credibilidad (Z): {Z:.2f}")
print(f"Prima creíble: {prima_creible:.2f}")
```

```python
# Visualización de las medias y la prima creíble
plt.figure(figsize=(15, 5))
plt.bar(['Media Individual', 'Media Mercado', 'Prima Creíble'], 
        [media_individual, media_mercado, prima_creible], color=['blue', 'lightblue', 'darkblue'], width= 0.3)
plt.ylabel("Valor de la Prima")
plt.title("Cálculo de Prima Creíble con Teoría de la Credibilidad")
plt.show()
```

# Credibilidad 3

```python
data = pd.read_excel("Ejemplo sieniestros.xlsx")
```

```python
data.head()
```

```python
# Calcular la media de los siniestros individuales y del mercado
media_individual = data['Datos Individuales'].mean()
media_mercado = data['Datos del Mercado'].mean()
```

```python
# Calcular las varianzas de los datos individuales y del mercado
var_individual = data['Datos Individuales'].var()
var_mercado = data['Datos del Mercado'].var()
```

```python
# Definir el número de datos individuales
n = len(data['Datos Individuales'])  # Número de observaciones individuales
```

```python
# Calcular el parámetro de credibilidad λ (Bayes)
lambda_bayes = var_individual / var_mercado  # Parámetro bayesiano basado en las varianzas
```

```python
# Calcular la prima creíble usando la fórmula bayesiana
prima_creible_bayes = (n / (n + lambda_bayes)) * media_individual + (lambda_bayes / (n + lambda_bayes)) * media_mercado
```

```python
# Imprimir los resultados
print(f"Media de siniestros individual: {media_individual:.2f}")
print(f"Media de siniestros del mercado: {media_mercado:.2f}")
print(f"Varianza de los siniestros individuales: {var_individual:.2f}")
print(f"Varianza de los siniestros del mercado: {var_mercado:.2f}")
print(f"Parámetro de credibilidad (λ): {lambda_bayes:.2f}")
print(f"Prima creíble (Bayes): {prima_creible_bayes:.2f}")
```

```python
# Visualización de las medias y la prima creíble
plt.figure(figsize=(15, 5))
plt.bar(['Media Individual', 'Media Mercado', 'Prima Creíble (Bayes)'], 
        [media_individual, media_mercado, prima_creible_bayes], color=['blue', 'lightblue', 'darkblue'], width= 0.3)
plt.ylabel("Valor de la Prima")
plt.title("Cálculo de Prima Creíble con Teoría Bayesiana")
plt.show()
```
